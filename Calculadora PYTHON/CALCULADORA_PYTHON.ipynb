```python
import tkinter as tk

# Colores personalizados
bg_color = "#F4F4F4"
btn_bg_color = "#D4D4D4"
btn_text_color = "#000000"
btn_highlight_bg_color = "#BFBFBF"
btn_highlight_text_color = "#FFFFFF"
display_bg_color = "#EDEDED"
display_text_color = "#000000"

def handle_button_click(text):
    if text == "=":
        try:
            result = eval(expression.get())
            expression.set(str(result))
            save_result(result)
        except:
            expression.set("Error")
    elif text == "C":
        expression.set("")
    elif text == "+/-":
        try:
            expression.set(str(-1 * eval(expression.get())))
        except:
            expression.set("Error")
    elif text == "%":
        try:
            expression.set(str(eval(expression.get()) / 100))
        except:
            expression.set("Error")
    elif text == "MR":
        expression.set(str(get_saved_result()))
    else:
        expression.set(expression.get() + text)

def create_rounded_button(canvas, x, y, radius, text):
    button = tk.Button(canvas, text=text, font=("Arial", 20), bd=0, bg=btn_bg_color, fg=btn_text_color, activebackground=btn_highlight_bg_color, activeforeground=btn_highlight_text_color)
    button.place(x=x, y=y, width=2*radius, height=2*radius, anchor="center")
    button.configure(command=lambda text=text: handle_button_click(text))

    canvas.create_oval(x - radius, y - radius, x + radius, y + radius, fill=btn_bg_color, outline=btn_bg_color)
    canvas.create_text(x, y, text=text, font=("Arial", 20), fill=btn_text_color)

def save_result(result):
    global saved_result
    saved_result = result

def get_saved_result():
    global saved_result
    return saved_result

root = tk.Tk()
root.title("Calculator")
root.configure(bg=bg_color)

expression = tk.StringVar()

entry = tk.Entry(root, textvariable=expression, font=("Arial", 24), justify="right", bg=display_bg_color, fg=display_text_color, bd=0)
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10, sticky="we")

buttons = [
    ("7", 50, 120), ("8", 150, 120), ("9", 250, 120), ("/", 350, 120),
    ("4", 50, 220), ("5", 150, 220), ("6", 250, 220), ("*", 350, 220),
    ("1", 50, 320), ("2", 150, 320), ("3", 250, 320), ("-", 350, 320),
    ("C", 50, 420), ("0", 150, 420), (".", 250, 420), ("+", 350, 420),
    ("+/-", 50, 520), ("%", 150, 520), ("=", 250, 520), ("MR", 350, 520)
]

canvas_width = 400
canvas_height = 620
canvas = tk.Canvas(root, width=canvas_width, height=canvas_height, bg=bg_color, bd=0, highlightthickness=0)
canvas.grid(row=1, column=0, columnspan=4, padx=10, pady=10)

button_radius = 40

for button_text, button_x, button_y in buttons:
    create_rounded_button(canvas, button_x, button_y, button_radius, button_text)

root.mainloop()
```
